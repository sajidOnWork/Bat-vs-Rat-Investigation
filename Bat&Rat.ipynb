```python
d1 = pd.read_csv("dataset1.csv")
d2 = pd.read_csv("dataset2.csv")
```

```python
OUT_DIR = "/content/outputs"
os.makedirs(OUT_DIR, exist_ok=True)
```

```python
plt.figure(figsize=(6,6))
d1['season'].value_counts(dropna=False).plot(kind='pie', autopct='%.1f')
plt.title('Distribution of Season (dataset1)')
plt.ylabel("")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "d1_pie_season.png"), dpi=200)
plt.show()
```

```python
plt.figure(figsize=(6,6))
d1['month'].value_counts(dropna=False).plot(kind='pie', autopct='%.1f')
plt.title('Distribution of Month (dataset1)')
plt.ylabel("")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "d1_pie_month.png"), dpi=200)
plt.show()
```

```python
plt.figure(figsize=(6,6))
d1['risk'].value_counts(dropna=False).plot(kind='pie', autopct='%.1f')
plt.title('Distribution of Risk (dataset1)')
plt.ylabel("")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "d1_pie_risk.png"), dpi=200)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='month', y='bat_landing_to_food', data=d1, cut=0)
plt.title('Bat Landing to Food by Month (dataset1)')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "d1_violin_bat_landing_to_food_by_month.png"), dpi=200)
plt.show()
```

```python
plt.figure(figsize=(6,6))
d2['month'].value_counts(dropna=False).plot(kind='pie', autopct='%.1f')
plt.title('Distribution of Month (dataset2)')
plt.ylabel("")
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "d2_pie_month.png"), dpi=200)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(d2['bat_landing_number'], kde=True, bins=30)
plt.title('Distribution of Bat Landing Number (dataset2)')
plt.xlabel('Bat Landing Number')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig(os.path.join(OUT_DIR, "d2_hist_bat_landing_number.png"), dpi=200)
plt.show()
```
